# Setup

## Local

* Install Python: Python3 using `Anaconda` environment
* Install Python Packages:

```python
import importlib.util
import os

package_name = ['nibabel', 'nilearn', 'pybids']

for i in package_name:
    spec = importlib.util.find_spec(i)
    if spec is None:
        print(i + ' is not installed.')
        os.system('pip install ' + i)
    else:
        print(i + ' is installed.')
```

* Install `dcm2niix`: We will be using the `dcm2niix` software for converting neuroimaging data from the DICOM format to NIfTI format.

```python
i = 'dcm2niix'
os.system('conda list ' + i)
os.system('conda install -c conda-forge ' + i)
print(i + ' is installed.')
```

* Install `DataLad`: A tool for managing and version controlling large datasets.

```python
package_name = ['datalad']

for i in package_name:
    spec = importlib.util.find_spec(i)
    if spec is None:
        print(i + ' is not installed.')
        os.system('conda install -c conda-forge ' + i)
    else:
        print(i + ' is installed.')
```

* Obtain lesson material: Using `DataLad` to download the [ds000030 dataset](https://openneuro.org/datasets/ds000030/versions/1.0.0) found on OpenNeuro.

```python
if not os.path.isdir('data'):
    os.makedirs('data')

os.system('datalad clone https://github.com/OpenNeuroDatasets/ds000030.git data/dc-mri/data')
```
